<h2>Drop the least correlated

```python
data = data.drop(['UsingPopupWindow','Favicon','IframeRedirection'],axis=1)
```

```python
data.head()
```

```python
X = data.drop('class',axis=1)
```

```python
y = data['class'].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

<h2>LogisticRegression

```python
log_model = LogisticRegression()
log_model.fit(X_train,y_train)
```

```python
y_preds = log_model.predict(X_test)
```

```python
print(accuracy_score(y_test,y_preds))
```

```python
print(confusion_matrix(y_test,y_preds))
```

```python
print(classification_report(y_test,y_preds))
```

<h2>DecisionTree

```python
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
y_preds = dt.predict(X_test)
```

```python
print(accuracy_score(y_test,y_preds))
```

```python
print(confusion_matrix(y_test,y_preds))
```

```python
print(classification_report(y_test,y_preds))
```

<h2>RAndomForestClassification with hyperparameter tuning

```python
from sklearn.model_selection import RandomizedSearchCV
import numpy as np
```

```python
n_estimators =[int(x) for x in np.linspace(start= 100,stop=500,num=10)]
criterion = ['gini', 'entropy']
max_features = ['auto', 'sqrt']
min_samples_split = [2, 5, 10]

random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'criterion': criterion,
               'min_samples_split': min_samples_split,
    
}
```

```python
rf = RandomForestClassifier()
rf_random = RandomizedSearchCV(rf, param_distributions=random_grid, cv=5, n_jobs=-1)
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_params_
```

```python
newrf_random = RandomForestClassifier(n_estimators=366,
 min_samples_split=2,
 max_features='auto',
 criterion ='gini')
newrf_random.fit(X_train,y_train)
```

```python
y_preds = newrf_random.predict(X_test)
```

```python
accuracy_score(y_test,y_preds)
```
